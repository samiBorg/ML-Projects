```python
import os
import PyPDF2
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


```

```python
# Set the path for the resumes folder and the job description file
resumes_folder_path = "/content/media"
job_description_path = "/content/drive/MyDrive/Project_data/Campus_JD_2023.pdf"

# Read the job description PDF file
job_description_file = open(job_description_path, 'rb')
job_description_reader = PyPDF2.PdfReader(job_description_file)
job_description = ""
for page in range(len(job_description_reader.pages)):
    job_description += job_description_reader.pages[page].extract_text()

# Tokenize the job description
stop_words = set(stopwords.words('english'))
job_description_tokens = nltk.word_tokenize(job_description)
job_description_tokens = [token.lower() for token in job_description_tokens if token.isalpha() and token.lower() not in stop_words]

# Create the TF-IDF vectorizer for the job description
vectorizer = TfidfVectorizer()
job_description_vector = vectorizer.fit_transform([job_description])

# Create a dictionary to store the results
results = {}

# Loop through the resumes folder
for filename in os.listdir(resumes_folder_path):
    # Check if the file is a PDF
    if filename.endswith('.pdf'):
        # Read the resume PDF file
        resume_file = open(os.path.join(resumes_folder_path, filename), 'rb')
        resume_reader = PyPDF2.PdfReader(resume_file)
        resume = ""
        for page in range(len(resume_reader.pages)):
            resume += resume_reader.pages[page].extract_text()

        # Tokenize the resume
        resume_tokens = nltk.word_tokenize(resume)
        resume_tokens = [token.lower() for token in resume_tokens if token.isalpha() and token.lower() not in stop_words]

        # Create the TF-IDF vectorizer for the resume
        resume_vector = vectorizer.transform([resume])

        # Calculate the cosine similarity between the job description and the resume
        cosine_sim = cosine_similarity(job_description_vector, resume_vector)[0][0]

        # Find the matched keywords between the job description and the resume
        matched_keywords = set(job_description_tokens).intersection(set(resume_tokens))

        # Store the results in the dictionary
        results[filename] = {
            'cosine_similarity': cosine_sim,
            'matched_keywords': matched_keywords
        }

f = open("sam.txt", "w")
# Print the results in the descending order of the cosine similarity percentages
for filename, data in sorted(results.items(), key=lambda x: x[1]['cosine_similarity'], reverse=True):
    print("Resume:", filename)
    print("Cosine Similarity:", data['cosine_similarity'])
    print("Matched Keywords:", data['matched_keywords'])
    print("\n")


    f.write("\nResume:")
    f.write(filename)
    f.write("\nCosine Similarity:")
    a=str(data['cosine_similarity']*100)+"%"
    f.write(a)
    f.write("\nMatched Keywords:")
    f.write(str(data['matched_keywords']))
    f.write("\n")

f.close()
```
